```python
import multiprocessing

# Function to download a video given its URL
def download_video(url, output_folder):
    try:
        yt = YouTube(url)
        video_stream = yt.streams.get_highest_resolution()
        print(f"Downloading: {yt.title}...")
        video_stream.download(output_folder)
        print(f"Download complete for {yt.title}!\n")
    except Exception as e:
        print(f"Error downloading video from {url}: {e}\n")

# Path to the file containing YouTube video URLs
file_path = '/content/drive/MyDrive/Flipkart/flipkart_scrap.csv'
output_folder = '/content/drive/MyDrive/Flipkart/flipkart_Keyframes|'

# Read video URLs from the file using 'latin-1' encoding
with open(file_path, 'r', encoding='latin-1') as file:
    video_urls = file.readlines()

# Remove whitespace characters such as '\n' from each URL and skip empty lines
video_urls = [url.strip() for url in video_urls if url.strip()]

# Number of processes to use
num_processes = 10

# Create and start processes
with multiprocessing.Pool(processes=num_processes) as pool:
    pool.starmap(download_video, [(url, output_folder) for url in video_urls])

print("All downloads completed successfully!")
```

```python
# Replace the csv_file_path with the path to your CSV file containing YouTube video URLs
csv_file_path = r"/content/drive/MyDrive/Work file/youtube video download/My11circle.csv"   # Scrap csv
df_video_urls = pd.read_csv(csv_file_path, encoding='latin1')

video_urls = df_video_urls['Video URL'].tolist()

video_data = []

for video_url in video_urls:
    try:
        print(f"Fetching data for video URL: {video_url}")

        # Extract video ID from the URL
        video_id = video_url.split("v=")[1]

        # Get video details from the YouTube API
        video_info = youtube.videos().list(part='snippet,contentDetails,statistics', id=video_id).execute()
        video_details = video_info['items'][0]

        # Convert duration to a more human-readable format
        duration_iso8601 = video_details['contentDetails']['duration']
        video_duration = isodate.parse_duration(duration_iso8601)

        video_data.append({
            'Video URL': video_url,
            'Video Title': video_details['snippet']['title'],
            'Channel Title': video_details['snippet']['channelTitle'],
            'Publish Time': video_details['snippet']['publishedAt'],
            'Views': video_details['statistics']['viewCount'],
            'Likes': video_details['statistics'].get('likeCount', 0),
            'Dislikes': video_details['statistics'].get('dislikeCount', 0),
            'Comments': video_details['statistics'].get('commentCount', 0),
            'Video Duration': str(video_duration),
            'Video Description': video_details['snippet'].get('description', ''),
        })

    except Exception as e:
        print(f"Error fetching data for video URL: {video_url}")
        print(f"Error details: {str(e)}")

# Create DataFrame only if video_data is not empty
if video_data:
    df_video_data = pd.DataFrame(video_data)
else:
    print("No video details available.")

# Save the final DataFrame to a CSV file
df_video_data.to_csv("My11circle_Final.csv", index=False)   # output csv meta data csv

# Display the final DataFrame
print(df_video_data)
```

```python
import os
from moviepy.editor import VideoFileClip
import csv

def get_video_metadata(file_path):
    clip = VideoFileClip(file_path)
    file_name = os.path.basename(file_path)
    file_name_without_extension = os.path.splitext(file_name)[0]  # Remove file extension
    return {
        'file_name': file_name_without_extension,
        'duration': clip.duration,
        'resolution': clip.size,
        # Add more metadata fields as needed
    }

def create_csv(folder_path, csv_path):
    with open(csv_path, 'w', newline='') as csvfile:
        fieldnames = ['file_name', 'duration', 'resolution']
        # Add more fieldnames as needed
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for file_name in os.listdir(folder_path):
            if file_name.endswith('.mp4'):
                file_path = os.path.join(folder_path, file_name)
                video_metadata = get_video_metadata(file_path)
                writer.writerow(video_metadata)

folder_path = '/content/drive/MyDrive/Work file/youtube video download/My11circle'
csv_path = '/content/My11circle_Final.csv'

create_csv(folder_path, csv_path)
```
